### Outliers

```python
def find_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-1.5*IQR
    high_val=q3+1.5*IQR
    print("Low:",low_val,"High:",high_val)
    ds=ds.loc[(ds[col]<low_val)|(ds[col]>high_val)]
    return ds
```

```python
import pandas as pd
df=pd.read_csv("auto-mpg.csv")
find_outliers(df,'mpg')
```

```python
import pandas as pd
df=pd.read_csv("auto-mpg.csv")
find_outliers(df,'acceleration')
```

```python
def remove_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-1.5*IQR
    high_val=q3+1.5*IQR
    print("Low:",low_val,"High:",high_val)
    ds=ds.loc[(ds[col]>=low_val)&(ds[col]<=high_val)]
    return ds
```

```python
import pandas as pd
df=pd.read_csv("auto-mpg.csv")
remove_outliers(df,'acceleration')
```
