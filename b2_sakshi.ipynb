#### set_index()
- methods used to set one or more columns of a dataframe as the index.
- useful when we need to modify or add new indices to our data as it enhances data retrieval, indexing and merging tasks.

```python
import pandas as pd
data={'Name':["Alice","Bob","Charlie"],'Age':[25,30,35],
     'City':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data)
df
```

```python
df.set_index('Name')
```

```python
df
```

```python
df.set_index('Name',drop=False,inplace=True)
```

```python
df
```

```python
df1=df.set_index('Name',drop=False)
```

```python
df.set_index(['Name','Age'])
```

#### reset_index()
- used to manage and reset index of a dataframe.

```python
import pandas as pd
data={'Name':["Alice","Bob","Charlie"],'Age':[25,30,35],
     'City':['New York','Los Angeles','Chicago']}
df=pd.DataFrame(data,index=['a','b','c'])
df
```

```python
df_reset=df.reset_index(drop=True)
df_reset
```

#### unique() / nunique()

```python
import pandas as pd
data=[[10,20,0],[10,10,10],[10,20,30]]
df=pd.DataFrame(data)
df
```

```python
print(df.nunique())
```

```python
print(df[1].unique())
```
