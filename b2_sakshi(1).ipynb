#### Pandas Series
    ---> pandas series is like a column in a table 
    ---> it is a 1-D array holding data of any type

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=["X","Y","Z"])
print(myvar)
print(type(myvar))
print(myvar[1])
```

```python
import pandas as pd
a=(1.4,2,3,4,5)
myvar=pd.Series(a)
print(myvar)
```
